```python
import pandas as pd
import numpy as np

# Load existing CSV (with 2 columns)
df = pd.read_csv("data/demand_large.csv")

# Recreate temperature column
np.random.seed(42)
rows = len(df)
time = np.arange(1, rows + 1)

temperature = 28 + 5 * np.sin(time / 50) + np.random.normal(0, 1, rows)

# Add temperature column
df["temperature"] = temperature.round(2)

# Save updated CSV (overwrite file)
df.to_csv("data/demand_large.csv", index=False)

print("✅ demand_large.csv successfully updated to 3 columns")
print(df.head())  # shows the changes
```
